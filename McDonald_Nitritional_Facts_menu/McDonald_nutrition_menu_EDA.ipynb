```python
df_menu.rename(columns={'Total Fat (% Daily Value)':'Total_fat_D%','Saturated Fat (% Daily Value)':'Saturated_fat_D%',\
                        'Cholesterol (% Daily Value)':'Cholesterol_D%','Sodium (% Daily Value)':'Sodum_D%',\
                        'Carbohydrates (% Daily Value)':'Carbohydrates_D%','Dietary Fiber (% Daily Value)':'Dietary_fiber_D%',\
                        'Vitamin A (% Daily Value)':'Vitamin_A_D%','Vitamin C (% Daily Value)':'Vitamin_C_D%',\
                        'Calcium (% Daily Value)':'Calcium_D%','Iron (% Daily Value)':'Iron_D%'}, inplace=True)
df_menu[10:30]
```

```python
df_menu.describe()
```

The descriptive statistical analysis above shows that the average McDonald's Value meal contains approximately 368 Calories.

```python
#df_menu.query('Category == "Beverages"')
Calories_by_Categories = df_menu.groupby(['Category'])['Calories'].sum().reset_index()
Calories_by_Categories
```

```python
df_menu.Calories.sum()
```

```python
plt.pie(Calories_by_Categories['Calories'], labels= Calories_by_Categories['Category'], autopct= '%1.1f%%', startangle=90)
plt.title ('Calories Percentage by Catergory')
plt.show()
```

From our piechart we understand that Salads, Desserts and Beverages contibute least to the overall calories with percentage values 1.7%, 1.6% and 3.2% respectively.

```python
Grilled_chicken_sandwish = df_menu[df_menu['Item'].str.contains('Grilled Chicken')]
Grilled_mean = Grilled_chicken_sandwish.describe().mean()
Grilled_mean= pd.DataFrame(Grilled_mean).rename_axis(index='Nutrients').rename(columns={0: 'Mean'})
Grilled_mean
```

```python
Crispy_chicken_sandwish = df_menu[df_menu['Item'].str.contains('Crispy Chicken')]
Crispy_mean = Crispy_chicken_sandwish.describe().mean()
Crispy_mean=pd.DataFrame(Crispy_mean).rename_axis(index='Nutrients').rename(columns={0:'Mean'})
Crispy_mean
#, {0:'Mean'})
```

```python
plt.figure(figsize=(13,5))
plt.subplot(1,2,2)
#Plotting Bar of mean vs Grilled Chicken nutrients 
sns.barplot(data=Crispy_mean, x='Nutrients', y='Mean')
plt.title('Crispy Chicken Sandwish Nutrients')
plt.xticks(rotation=90)
plt.subplot(1,2,1)
sns.barplot(Grilled_mean, x = 'Nutrients', y = 'Mean', color='Orange')
plt.title('Grilled Chicken Sandwish Nutrients')
plt.xticks(rotation=90)
plt.suptitle('Bar Comparison of Grilled Chicken Sandwish Nutrients to Crispy Chicken Sandwish')
plt.show()
```

INSIGHTS:
- From our analysis of Grilled and Crispy Chicken we can understand that though both sandwishes have similar amounts of nutrients, crispy chicken sandwishes have significantly more sodium and calories than the grilled chicken sandwishes but generally the Grilled chicken sandwishes have more nutritional value that the crispy chicken sandwishes.
